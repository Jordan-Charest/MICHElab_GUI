```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def return_dataframe(filename):
    return pd.read_csv(filename).set_index("id")

df_names = ["WT"]
title_strings = ["WT mice"]
```

```python
for i in range(len(df_names)):

    df_name = df_names[i]
    title_str = title_strings[i]

    df = return_dataframe(f"WT_tests_2/nvc_{df_name}.csv")
    df_melted = pd.melt(df, id_vars=["number"], value_vars=['nvc_segment_0/2', 'nvc_segment_1/2', "nvc_segment_0/2_no_mvmt", "nvc_segment_1/2_no_mvmt"])

    # print(df_melted)

    ax = sns.barplot(x="number", y="value", hue="variable", data=df_melted)
    plt.title(f"Neurovascular coupling\n{title_str}")
    plt.show()
```

```python
for i in range(len(df_names)):

    df_name = df_names[i]
    title_str = title_strings[i]

    df = return_dataframe(f"WT_tests_2/funcsim_{df_name}.csv")
    df_melted = pd.melt(df, id_vars=["number"], value_vars=['funcsim_segment_0/2', 'funcsim_segment_1/2', 'funcsim_segment_0/2_no_mvmt', 'funcsim_segment_1/2_no_mvmt'])

    # print(df_melted)

    ax = sns.barplot(x="number", y="value", hue="variable", data=df_melted)
    plt.title(f"Functional similarity\n{title_str}")
    plt.show()
```
