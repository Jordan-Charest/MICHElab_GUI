Select the mice to add, the file identifier and the file path.

```python
# mice_num = ["39-12","42-12","44-12","45-12","46-12","251-6","254-6", "308-8", "308-10", "308-12", "308-14"]
mice_num = ["316-8","316-10","316-12","322-6","322-8","322-10","322-12","353-6","353-8",
            "353-10","361-6","365-6","367-6","374-6","374-8","374-10","387-6","387-10",
            "396-6","397-6","410-6","410-8","410-10","412-8","412-10","415-6","415-8"]


input_file_id = "v1_mvmt"
output_file = f"D:/mouse_data/new_data/df_output/GCaMP/GCaMP_{input_file_id}.csv"
# mice_num = mice_num[:4]

def return_filename(mouse_num, output_file_id): # Modify as needed according to your file structure
    return f"D:/mouse_data/new_data/M{mouse_num}/formatted/M{mouse_num}_{output_file_id}.h5"
```

Run this cell to create and print the dataframe.

```python
from toolbox_jocha.hdf5 import get_attributes
from toolbox_jocha.parsing import smart_cast
import pandas as pd


dataframe_dict = {"id": [], "number": []}
index = None

index = []

for mouse_num in mice_num:

    attr = get_attributes(return_filename(mouse_num, input_file_id), "data")

    # index.append(f"M{mouse_num}_{output_file_id}")
    dataframe_dict["number"].append(mouse_num.split("-")[0])
    dataframe_dict["id"].append(f"M{mouse_num}_{input_file_id}")

    for key, value in attr.items():

        try:
            dataframe_dict[key].append(smart_cast(value))
        except KeyError:
            dataframe_dict[key] = []
            dataframe_dict[key].append(smart_cast(value))


df = pd.DataFrame(dataframe_dict).set_index("id")

print(df)
```

Run this cell to save the dataframe to a csv file.

```python
df.to_csv(output_file)
```
