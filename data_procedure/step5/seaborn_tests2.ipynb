```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def return_dataframe(filename):
    return pd.read_csv(filename).set_index("id")
```

```python
df = return_dataframe("nvc_v1_df.csv")


# Create groups for plotting
groups = {
    "GCaMP": df.index[df["GCaMP"]],
    "Female": df.index[df['sex'] == 'F'],
    "Male": df.index[df['sex'] == 'M']
}
```

```python
df = return_dataframe("nvc_v1_df.csv")
ax = sns.barplot(x="GCaMP", y="nvc_segment_0/2", hue='age', data=df)
plt.title("Neurovascular Coupling")
plt.show()



df = return_dataframe("funcsim_v1_df.csv")
ax = sns.barplot(x="GCaMP", y="funcsim_segment_0/2", hue='age', data=df)
plt.title("Functional Similarity")
plt.show()




df = return_dataframe("funcrep_v1_df.csv")
ax = sns.barplot(x="GCaMP", y="GCaMP_funcrep_segment_0/2", hue='age', data=df)
plt.title("GCaMP Functional Representativity")
plt.show()

ax = sns.barplot(x="GCaMP", y="dHbT_funcrep_segment_0/2", hue='age', data=df)
plt.title("dHbT Functional Representativity")
plt.show()




df = return_dataframe("modularity_v1_df.csv")
ax = sns.barplot(x="GCaMP", y="GCaMP_modularity_segment_0/2", hue='age', data=df)
plt.title("GCaMP Modularity")
plt.show()

ax = sns.barplot(x="GCaMP", y="dHbT_modularity_segment_0/2", hue='age', data=df)
plt.title("dHbT Modularity")
plt.show()
```

```python
df = return_dataframe("modularity_v1_df.csv")
ax = sns.barplot(x="number", y="GCaMP_modularity_whole", hue='age', data=df)
plt.title("GCaMP Modularity")
plt.show()

ax = sns.barplot(x="age", y="dHbT_modularity_whole", hue='sex', data=df)
plt.title("dHbT Modularity")
plt.show()
```
