```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def return_dataframe(filename):
    return pd.read_csv(filename).set_index("id")
```

```python
df = return_dataframe("nvc_WT_df.csv")
df_melted = pd.melt(df, id_vars=["number"], value_vars=['nvc_segment_0/2', 'nvc_segment_1/2'])

# print(df_melted)

ax = sns.barplot(x="number", y="value", hue="variable", data=df_melted)
plt.title("Neurovascular coupling (WT mice)")
plt.show()

ax = sns.barplot(x="variable", y="value", data=df_melted)
plt.title("Neurovascular coupling (WT, all mice)")
plt.show()
```
