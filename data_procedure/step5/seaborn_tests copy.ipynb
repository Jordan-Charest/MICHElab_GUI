```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def return_dataframe(filename):
    return pd.read_csv(filename)
```

```python
df = return_dataframe("nvc_filter_df.csv")
df_long = pd.melt(df, id_vars='id', value_vars=['nvc_segment_0/2', 'nvc_segment_1/2'],
                  var_name='segment', value_name='nvc')

plt.figure(figsize=(12,5))
ax = sns.barplot(x="id", y="nvc", data=df_long, hue="segment")
plt.title("Neurovascular Coupling")
plt.show()


df_long = pd.melt(df, id_vars='id', value_vars=['lag_segment_0/2', 'lag_segment_1/2'],
                  var_name='segment', value_name='lag')
plt.figure(figsize=(12,5))
ax = sns.barplot(x="id", y="lag", data=df_long, hue="segment")
plt.title("lag")
plt.show()

df = return_dataframe("funcsim_filter_df.csv")
df_long = pd.melt(df, id_vars='id', value_vars=['funcsim_segment_0/2', 'funcsim_segment_1/2'],
                  var_name='segment', value_name='funcsim')
plt.figure(figsize=(12,5))
ax = sns.barplot(x="id", y="funcsim", data=df_long, hue="segment")
plt.title("Funcsim")
plt.show()


df = return_dataframe("funcrep_filter_df.csv")
df_long = pd.melt(df, id_vars='id', value_vars=['dHbT_funcrep_segment_0/2', 'dHbT_funcrep_segment_1/2'],
                  var_name='segment', value_name='dHbT_funcrep')
plt.figure(figsize=(12,5))
ax = sns.barplot(x="id", y="dHbT_funcrep", data=df_long, hue="segment")
plt.title("dHbT Funcrep")


df_long = pd.melt(df, id_vars='id', value_vars=['GCaMP_funcrep_segment_0/2', 'GCaMP_funcrep_segment_1/2'],
                  var_name='segment', value_name='GCaMP_funcrep')
plt.figure(figsize=(12,5))
ax = sns.barplot(x="id", y="GCaMP_funcrep", data=df_long, hue="segment")
plt.title("GCaMP Funcrep")
```
